<h3>HANDLING MISSING VALUES</h3>

```python
import pandas as pd

df = pd.read_csv("hospital_patient_data.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
sns.heatmap(df.isnull(), cbar=False, cmap='Reds')
plt.title("Missing Values Before Cleaning")
plt.show()
```

```python
# Median is more robust to outliers
median_age = df["Age"].median()
df["Age"] = df["Age"].fillna(median_age)

# Mode gives the most frequent diagnosis
mode_diag = df["Diagnosis"].mode()[0]
df["Diagnosis"] = df["Diagnosis"].fillna(mode_diag)

# Fill missing treatment cost using mean per department
df["TreatmentCost"] = df.groupby("Department")["TreatmentCost"].transform(lambda x: x.fillna(x.mean()))
```

```python
df.isnull().sum()
```

```python
# Save cleaned data
df.to_csv("hospital_patient_data_cleaned.csv", index=False)
```

<h3>EDA- EXPLORATORY DATA ANALYSYS</h3>

```python
# Patient distribution by department:
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 4))
sns.countplot(data=df, x="Department", order=df["Department"].value_counts().index)
plt.title("Number of Patients by Department")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.barplot(data=df, x="Diagnosis", y="LengthOfStay", estimator="mean")
plt.title("Average Length of Stay by Diagnosis")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Readmission rate by department:
readmit_dept = df.groupby("Department")["Readmitted"].value_counts(normalize=True).unstack()
readmit_dept["Yes"].plot(kind='bar', figsize=(8, 4), title="Readmission Rate by Department")
plt.ylabel("Proportion Readmitted")
plt.tight_layout()
plt.show()
```

```python
# Age distribution of readmitted patients:
plt.figure(figsize=(8, 4))
sns.histplot(data=df[df["Readmitted"] == "Yes"], x="Age", bins=15, kde=True)
plt.title("Age Distribution of Readmitted Patients")
plt.tight_layout()
plt.show()
```

```python
# Avg treatment cost by diagnosis:
plt.figure(figsize=(8, 4))
sns.barplot(data=df, x="Diagnosis", y="TreatmentCost", estimator="mean")
plt.title("Average Treatment Cost by Diagnosis")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

<h3> EDA: Beyond the Basics</h3>

```python
# 1. Correlation Heatmap (Numerical Variables)
# To See how numerical features like Age, Length of Stay, and Cost relate to each other.
plt.figure(figsize=(6, 4))
sns.heatmap(df[["Age", "LengthOfStay", "TreatmentCost"]].corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.tight_layout()
plt.show()
```

```python
# 2. Boxplot: Treatment Cost by Department
# Spot outliers or cost variations by department.
plt.figure(figsize=(8, 4))
sns.boxplot(data=df, x="Department", y="TreatmentCost")
plt.title("Treatment Cost Distribution by Department")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 3. Pivot Table: Avg Stay & Readmission by Department
pivot = df.pivot_table(
    index="Department",
    values=["LengthOfStay", "TreatmentCost"],
    aggfunc="mean"
)

readmit_rate = df[df["Readmitted"] == "Yes"].groupby("Department").size() / df.groupby("Department").size()
pivot["ReadmissionRate"] = readmit_rate

pivot.sort_values("ReadmissionRate", ascending=False)
```

<h3>SQL</h3>

```python
import pandas as pd
import sqlite3

# Load your CSV file
df = pd.read_csv("hospital_patient_data.csv")  # use correct filename

# Connect or create SQLite database
conn = sqlite3.connect("hospital_db.db")

# Write the DataFrame into SQLite as a table
df.to_sql("patients", conn, if_exists="replace", index=False)

# Preview to confirm
df.head()
```

```python
# 1. Total Patients by Department
query = '''
SELECT Department, COUNT(*) AS TotalPatients
FROM patients
GROUP BY Department
ORDER BY TotalPatients DESC;
'''
pd.read_sql_query(query, conn)
```

```python
# 2. Average Length of Stay by Diagnosis
query = '''
SELECT Diagnosis, ROUND(AVG(LengthOfStay), 2) AS AvgStay
FROM patients
GROUP BY Diagnosis
ORDER BY AvgStay DESC;
'''
pd.read_sql_query(query, conn)
```

```python
# 3. Readmission Rate by Department
query = '''
SELECT 
    Department,
    ROUND(SUM(CASE WHEN Readmitted = 'Yes' THEN 1 ELSE 0 END) * 100.0 / COUNT(*), 2) AS ReadmissionRatePercent
FROM patients
GROUP BY Department
ORDER BY ReadmissionRatePercent DESC;
'''
pd.read_sql_query(query, conn)
```

```python
# 4. Average Treatment Cost by Department
query = '''
SELECT Department, ROUND(AVG(TreatmentCost), 2) AS AvgCost
FROM patients
GROUP BY Department
ORDER BY AvgCost DESC;
'''
pd.read_sql_query(query, conn)
```

```python
# 5. Top 5 Most Expensive Diagnoses (Avg Treatment Cost)
query = '''
SELECT Diagnosis, ROUND(AVG(TreatmentCost), 2) AS AvgCost
FROM patients
GROUP BY Diagnosis
ORDER BY AvgCost DESC
LIMIT 5;
'''
pd.read_sql_query(query, conn)
```
